```python
#imports here
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
# Set Chrome options to disable notifications
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")

# Initialize Chrome webdriver
driver = webdriver.Chrome(options=chrome_options)

# Open Facebook in the browser
driver.get("https://www.facebook.com")
```

```python
# Login to Facebook
username = driver.find_element(By.CSS_SELECTOR, "input[name='email']")
password = driver.find_element(By.CSS_SELECTOR, "input[name='pass']")

username.send_keys("email")
password.send_keys("passwrd")

driver.find_element(By.CSS_SELECTOR, "button[type='submit']").click()

# Wait for login to complete
time.sleep(5)

#We are logged in!
print("Logged in")
```

```python
# Navigate to the post from which you want to scrape comments
post_url = "https://www.facebook.com/watch/?v=954359099458540&__cft__[0]=AZVyoePQU2ejenNo1vfVECvo0GRN4Z9L0phdryZuxs5jPSMFsGi-lHLdTE5YU8Hyvgp521k6F_6cr9HY2RMdAhUbyMTUJZaWDLcK6LqKkw_LaZ5kFFiddvnVaZgSuL9bjlVbrWwmwndp6kwIMkcRPOWOVL_rDLmhZBrdI-Xot6pm7e1mCDNYbz9b6A5eOmVUgKY&__tn__=%2CO%2CP-R"
driver.get(post_url)
# Wait for the comments section to load
time.sleep(5)
```

```python
# Updated selector for "Load More Comments" button
load_more_button = driver.find_element(By.CSS_SELECTOR, ".xdt5ytf > .x9f619")
driver.execute_script("arguments[0].click();", load_more_button)
time.sleep(2) # Wait for comments to load

# Updated selector for scraping comments
comments = driver.find_elements(By.CSS_SELECTOR, ".x1lliihq > .x193iq5w  > .xdj266r > div")

# Extract comment text
comment_text = [comment.text for comment in comments]

# Close the browser
# driver.quit()

# Print the comment text
for comment in comment_text:
    print(comment + "\n")
    
```

```python
# Create a dataframe
data = pd.DataFrame({'Comment': comment_text})

# Save data to CSV
data.to_csv('Facebook_comments.csv', index=False)
```
